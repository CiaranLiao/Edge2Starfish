```python
from keras.models import load_model
from matplotlib import pyplot as plt

# 加载模型
model = load_model('starfish4w.h5')
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow import keras

# 加载模型结构的H5文件
model_structure_path = 'starfishT.h5'
model = keras.models.load_model(model_structure_path)

# 加载模型权重的H5文件
model_weights_path = 'starfishT_weights.h5'
model.load_weights(model_weights_path)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
def generate_images(model, test_input):
  # test_input=tf.expand_dims(test_input,axis=0)
  prediction = model(test_input, training=True)
  plt.figure(figsize=(15, 15))

  display_list = [test_input[0], prediction[0]]
  title = ['Input Image',  'Predicted Image']

  for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.title(title[i])
    # Getting the pixel values in the [0, 1] range to plot.
    plt.imshow(display_list[i] * 0.5 + 0.5)
    plt.axis('off')
  plt.show()
```

```python
import numpy as np
from PIL import Image
import tensorflow as tf

#image_path = input("图片文件路径：")
image_path="draw.jpg"
image = tf.io.read_file(image_path)

# 解码图片文件并将其转换为tf.Tensor对象
image = tf.image.decode_image(image, dtype=tf.float32, channels=3)
image = tf.expand_dims(image, axis=0)

generate_images(model, image)
```
